```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("/content/Covid Data.csv")
```

```python
df
```

```python
df_sorted = df.sort_values('AGE')
x = df_sorted['AGE']
y = df_sorted['ICU']
```

```python
df.info()
```

```python
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('AGE')
plt.ylabel('ICU')
plt.show()
```

```python
# Scatter Plot
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
mean_ages = df.groupby('CLASIFFICATION_FINAL')['AGE'].mean()

# Bar Plot
plt.bar(mean_ages.index, mean_ages)
plt.title('Mean Age by Classification')
plt.xlabel('Classification Final')
plt.ylabel('Mean Age')
plt.show()
```

```python
plt.hist(df['AGE'], bins=30, edgecolor='black')  # You can adjust the number of bins
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sex_counts = df['SEX'].value_counts()

# Pie Chart
plt.pie(sex_counts, labels=['Female', 'Male'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Patients by Sex')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['SEX'] == 0]['AGE'], df[df['SEX'] == 1]['AGE']], labels=['Female', 'Male'])
plt.title('Box Plot of Age by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
