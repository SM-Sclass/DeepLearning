```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/content/Covid Data.csv")
```

```python
#Countplot for categorical columns:
sns.countplot(data=df, x='SEX')
sns.countplot(data=df, x='PATIENT_TYPE')
sns.countplot(data=df, x='PNEUMONIA')
sns.countplot(data=df, x='DIABETES')
```

```python
#Histogram for continuous columns
sns.histplot(data=df, x='AGE', kde=True)
sns.histplot(data=df, x='MEDICAL_UNIT', kde=True)
```

```python
# Boxplot with categorical variables for comparison
sns.boxplot(data=df, x='SEX', y='AGE')
sns.boxplot(data=df, x='PATIENT_TYPE', y='AGE')
sns.boxplot(data=df, x='PNEUMONIA', y='AGE')
```

```python
#Boxplot for detecting outliers in continuous columns
sns.boxplot(data=df, y='AGE')
sns.boxplot(data=df, y='MEDICAL_UNIT')
```

```python
#Pairplot to analyze pairwise relationships
sns.pairplot(df[['AGE', 'SEX', 'PATIENT_TYPE', 'PNEUMONIA', 'DIABETES']])
```

```python
#Violinplot for comparing distributions
sns.violinplot(data=df, x='SEX', y='AGE')
sns.violinplot(data=df, x='DIABETES', y='AGE')
```

```python
#Barplot for aggregation
sns.barplot(data=df, x='SEX', y='AGE')
sns.barplot(data=df, x='PNEUMONIA', y='AGE')
```

```python
#FacetGrid for visualizing relationships across subsets
g = sns.FacetGrid(df, col='SEX')
g.map(sns.histplot, 'AGE')
```

```python
#Lineplot to check trends
sns.lineplot(data=df, x='AGE', y='MEDICAL_UNIT')
sns.lineplot(data=df, x='AGE', y='ICU')
```
